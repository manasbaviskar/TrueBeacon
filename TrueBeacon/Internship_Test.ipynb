### Extracting/Formatting the CSV files 

```python
index_df = pd.read_csv("Index_Subset.csv")
scores_df = pd.read_csv("Scores_Subset.csv")
scores_df['Unnamed: 0'] = pd.to_datetime(scores_df['Unnamed: 0'], format = '%Y-%m-%d')
scores_df = scores_df.set_index('Unnamed: 0')
scores_df.index  = scores_df.index.strftime("%d-%m-20%y")
index_df = index_df.set_index('Unnamed: 0')

cols = []
for i in scores_df.columns:
    #scores_df[i] = i
    cols.append(i)
    
a = scores_df.copy()
```

```python
index_df
```

### Creating a hashcode for each symbol in score column

```python
hashcode = {}
for c,i in enumerate(cols):
    hashcode[i] = c+1
```

```python
for i in index_df.columns:
    index_df[i]=index_df[i].map(hashcode)
```

```python
#Since we need top 200 symbols, we drop the redundant column i.e the last one
index_df = index_df.drop(['200'], axis = 1)
```

```python
index_df = index_df.fillna(0)
for i in index_df.columns:
    index_df[i]=index_df[i].astype(int)
```

```python
index_df
```

```python
%%timeit
for days in range(index_df.shape[0]):
    for i in range(index_df.shape[1]):
        if np.isnan(index_df.iloc[days,i]):
            continue
        else:
            a.iloc[days, int(index_df.iloc[days,i]-1)] = 1
```
